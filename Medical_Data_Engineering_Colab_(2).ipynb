### **[1] Import Libraries**

```python
import sqlite3
import pandas as pd
import time
```

### **[2] Load Excel Data**

```python
from google.colab import files
uploaded = files.upload()
```

```python
xls_path = "Updated_Medical_Appointments_Corrected.xlsx"

df_patients = pd.read_excel(xls_path, sheet_name="Patients")
df_appointments = pd.read_excel(xls_path, sheet_name="Appointments")
df_conditions = pd.read_excel(xls_path, sheet_name="MedicalConditions")
```

### **[3] Create Database**

```python
conn = sqlite3.connect("medical_project.db")
cursor = conn.cursor()
```

### **[4] Create Tables**

```python
cursor.execute("DROP TABLE IF EXISTS MedicalConditions")
cursor.execute("DROP TABLE IF EXISTS Appointments")
cursor.execute("DROP TABLE IF EXISTS Patients")

cursor.execute("""
CREATE TABLE Patients (
    PatientId TEXT PRIMARY KEY,
    Gender TEXT,
    Age INTEGER
);
""")

cursor.execute("""
CREATE TABLE Appointments (
    AppointmentId TEXT PRIMARY KEY,
    PatientId TEXT,
    ScheduledDay TEXT,
    AppointmentDay TEXT,
    SMS_received INTEGER,
    No_show TEXT,
    FOREIGN KEY (PatientId) REFERENCES Patients(PatientId)
);
""")

cursor.execute("""
CREATE TABLE MedicalConditions (
    ConditionId TEXT PRIMARY KEY,
    PatientId TEXT,
    Hypertension INTEGER,
    Diabetes INTEGER,
    Alcoholism INTEGER,
    Handcap INTEGER,
    FOREIGN KEY (PatientId) REFERENCES Patients(PatientId)
);
""")
```

### **[5] Insert Data into Tables**

```python
df_patients.to_sql("Patients", conn, if_exists="append", index=False)
df_appointments.to_sql("Appointments", conn, if_exists="append", index=False)
df_conditions.to_sql("MedicalConditions", conn, if_exists="append", index=False)
```

### **[9] SQL Query Tasks & Indexing**

```python
query1 = """
SELECT DISTINCT p.PatientId, p.Gender, a.No_show
FROM Patients p
JOIN Appointments a ON p.PatientId = a.PatientId
WHERE a.No_show = 'Yes'
"""
df1 = pd.read_sql_query(query1, conn)
print("Task 1:", df1)

query2 = """
SELECT p.PatientId, COUNT(*) AS total_sms
FROM Patients p
JOIN Appointments a ON p.PatientId = a.PatientId
WHERE a.SMS_received = 1
GROUP BY p.PatientId
"""
df2 = pd.read_sql_query(query2, conn)
print("Task 2:", df2)

start = time.time()
cursor.execute("SELECT * FROM Appointments WHERE No_show = 'Yes'")
print("Time without index:", time.time() - start)

cursor.execute("CREATE INDEX IF NOT EXISTS idx_no_show ON Appointments(No_show)")
start = time.time()
cursor.execute("SELECT * FROM Appointments WHERE No_show = 'Yes'")
print("Time with index:", time.time() - start)

query4 = """
SELECT p.PatientId, a.AppointmentDay
FROM Patients p
JOIN Appointments a ON p.PatientId = a.PatientId
WHERE a.AppointmentDay BETWEEN '2025-01-01' AND '2025-01-31'
"""
df4 = pd.read_sql_query(query4, conn)
print("Task 4:", df4)

start = time.time()
cursor.execute("SELECT * FROM Appointments WHERE AppointmentDay >= '2025-01-01'")
print("Time without date index:", time.time() - start)

cursor.execute("CREATE INDEX IF NOT EXISTS idx_appointment_day ON Appointments(AppointmentDay)")
start = time.time()
cursor.execute("SELECT * FROM Appointments WHERE AppointmentDay >= '2025-01-01'")
print("Time with date index:", time.time() - start)

conn.close()
```
